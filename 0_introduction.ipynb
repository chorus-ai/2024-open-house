# Introduction to CHoRUS waveforms

The CHoRUS dataset comprises structured EHR data integrated with images and waveforms. This notebook introduces electrocardiograms (ECGs), which offer important insights into patient health.

## Load sample data

```python
!pip install wfdb
import wfdb
```

```python
# Use the MIT-BIH arrhythmia database on PhysioNet
# https://physionet.org/content/mitdb/

physionet_project = "mitdb"
record_name = "100"
```

```python
help(wfdb.rdsamp)
```

```python
# Load the record from PhysioNet

record = wfdb.rdrecord(record_name,
                       sampto = 15000,
                       pn_dir=physionet_project)
display(record)
```

```python
# Plot the record

wfdb.plot_wfdb(record)
```

```python
# Read associated annotation file

annotation = wfdb.rdann(record_name,
                        'atr',
                        sampto = 15000,
                        pn_dir=physionet_project)
```

```python
# Plot the previous signal, now with annotations

wfdb.plot_wfdb(record=record, annotation=annotation,
               title='Record 100 from MIT-BIH Arrhythmia Database',
               time_units='seconds')
```
